```python
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import geopandas as gpd
import geoplot.crs as gcrs
import geoplot as gplt
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import fiona
```

```python
airbnbs = gpd.read_file('listings.csv')
```

```python
airbnbs.longitude = pd.to_numeric(airbnbs.longitude)
airbnbs.latitude = pd.to_numeric(airbnbs.latitude)
airbnbs = airbnbs.set_geometry(gpd.points_from_xy(airbnbs.longitude, airbnbs.latitude))
```

```python
airbnbs.price = airbnbs['price'].str.replace('$', '', regex = False)
airbnbs.price = airbnbs['price'].str.replace(',', '', regex = False)
airbnbs.price = pd.to_numeric(airbnbs.price)
```

```python
airbnbs.columns
```

```python
airbnbs.head()
```

```python
plt.boxplot(airbnbs.price)
```

```python
# Subset dataframe without outliers
airbnbs_no_outliers = airbnbs[(np.abs(stats.zscore(airbnbs['price'])) < 3)]
```

```python
plt.boxplot(airbnbs_no_outliers['price'])
```

```python
chi = gpd.read_file('chi_bounds.geojson')
ax = gplt.polyplot(chi, projection=gcrs.AlbersEqualArea(), edgecolor='white', facecolor='lightgray')
gplt.pointplot(airbnbs_no_outliers, ax=ax, hue = 'price', legend = True)
```

```python
chi
```

```python
airbnbs_by_nbh = dict()
airbnbs_no_outliers['neighborhood'] = np.nan
for index, airbnb in airbnbs_no_outliers.iterrows():
    point = airbnb['geometry']
    price = airbnb['price']
    for idx, nbh in chi.iterrows():
        if point.within(nbh['geometry']):
            if not (nbh['pri_neigh'] in airbnbs_by_nbh):
                airbnbs_by_nbh[nbh['pri_neigh']] = [price]
            else:
                airbnbs_by_nbh[nbh['pri_neigh']].append(price)
            airbnbs_no_outliers.at[index, 'neighborhood']= nbh['pri_neigh']
```

```python
airbnbs_no_outliers['neighborhood']
```

```python
chi['mean_price'] = 0
chi['median_price'] = 0
for i in range(chi.shape[0]):
    chi.at[i, 'mean_price'] = np.mean(airbnbs_by_nbh.get(chi.at[i, 'pri_neigh'], 0))
    chi.at[i, 'median_price'] = np.median(airbnbs_by_nbh.get(chi.at[i, 'pri_neigh'], 0))
```

```python
gplt.choropleth(chi, hue = 'mean_price', projection=gcrs.AlbersEqualArea(), cmap = 'inferno')
plt.savefig('mean_airbnbs_chi.png')
```

```python
gplt.choropleth(chi, hue = 'median_price', projection=gcrs.AlbersEqualArea(), cmap = 'inferno')
plt.savefig('median_airbnbs_chi.png')
```

```python
airbnbs_no_outliers['bathrooms'] = airbnbs_no_outliers['bathrooms_text'].str.split(' ').str[0]
```

```python
def bathroom_convert(input_str):
    try:
        return np.float64(input_str)
    except:
        if (input_str == ''):
            return np.nan
        else:
            return 0.5
airbnbs_no_outliers.bathrooms = airbnbs_no_outliers.bathrooms.apply(bathroom_convert)
```

```python
airbnbs_no_outliers.bedrooms = pd.to_numeric(airbnbs_no_outliers.bedrooms)
```

```python
airbnbs_no_outliers.review_scores_rating = pd.to_numeric(airbnbs_no_outliers.review_scores_rating)
```

```python
# Check missing values
pd.options.display.max_rows = 80
percent_missing = airbnbs_no_outliers.isna().sum() * 100/len(airbnbs_no_outliers)
miss_frame = pd.DataFrame({'Columns':airbnbs_no_outliers.columns, 'Missing Rate':percent_missing})
miss_frame
```

```python
airbnbs_no_outliers.dtypes
```

```python
# Create test and training sets
reg_frame = airbnbs_no_outliers.dropna()
X = reg_frame[['bedrooms', 'bathrooms', 'neighborhood', 'review_scores_rating']]
X = pd.get_dummies(X, columns=["neighborhood"])
y = reg_frame['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
reg = DecisionTreeRegressor(random_state = 42).fit(X_train, y_train)
```

```python
mean_squared_error(y_train, reg.predict(X_train))
```

```python
mean_squared_error(y_test, reg.predict(X_test))
```
